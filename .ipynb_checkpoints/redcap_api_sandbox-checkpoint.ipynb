```python
from requests import post

TOKEN = '6767D319D4AB8EC7B27D071C54C5657F'
URL = 'https://redcap.ucsf.edu/api/'

payload = {'token':TOKEN, 'format':'json', 'content':'metadata'}

response = post(URL, data=payload)

print(response.status_code)

data = response.json()

fields_histo = {}

for section in data:
    f_name = section['form_name']
    if f_name in fields_histo:
        fields_histo[f_name] += 1
    else:
        fields_histo[f_name] = 1
        
print(fields_histo)
    
```
